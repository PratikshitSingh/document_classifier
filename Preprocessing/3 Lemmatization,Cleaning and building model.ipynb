```python
import pandas as pd
import numpy as np
import re
import os
```

```python
def clean_str(string):
   
    string = re.sub(r"\'s", "", string)
    string = re.sub(r"\'ve", "", string)
    string = re.sub(r"n\'t", "", string)
    string = re.sub(r"\'re", "", string)
    string = re.sub(r"\'d", "", string)
    string = re.sub(r"\'ll", "", string)
    string = re.sub(r",", "", string)
    string = re.sub(r"!", " ! ", string)
    string = re.sub(r"\(", "", string)
    string = re.sub(r"\)", "", string)
    string = re.sub(r"\?", "", string)
    string = re.sub(r"'", "", string)
    string = re.sub(r"[^A-Za-z0-9(),!?\'\`]", " ", string)
    string = re.sub(r"[0-9]\w+|[0-9]","", string)
    string = re.sub(r"\s{2,}", " ", string)
    return string.strip().lower()
```

```python
dataset=pd.read_csv("../dataset/bbc/newdatasets.csv")
```

```python
x=dataset["news"].tolist()
y=dataset["type"].tolist()
```

```python
len(x)
```

```python
x
```

```python
for index,value in enumerate(x):
    x[index]=clean_str(value)
    
```

```python
x[0]
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

stop_words=set(stopwords.words('english'))

for index,value in enumerate(x):
    word_tokens=word_tokenize(x[index])
    f=[]
    for word in word_tokens:
        if word not in stop_words:
            f.append(word)
    new_string=""
    for w in f:
        new_string=new_string+w+" "
    x[index]=new_string[:]    
```

```python
x[0]
```

```python
import nltk
from nltk.stem import WordNetLemmatizer

wnl=WordNetLemmatizer()

for index,value in enumerate(x):
     string_lemma=""
     for w in x[index].split():
        string_lemma=string_lemma+wnl.lemmatize(w)+" "
     x[index]=string_lemma[:]   
```

```python
x[0]
```

```python
import scipy
from sklearn.feature_extraction.text import TfidfVectorizer

vect=TfidfVectorizer(min_df=2,stop_words="english")
```

```python
X=vect.fit_transform(x)
Y=np.array(y)
print("no of features extracted:")
eg, no_features = X.shape
print(no_features)
```

```python
from sklearn.model_selection import train_test_split

train_input, test_input, train_output, test_output = train_test_split(X, Y, test_size=0.2, random_state=25)

print("train input size:", train_input.shape)
print("test input size:", test_input.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=300, max_depth=150, n_jobs=1)
model.fit(train_input, train_output)
predicted_output = model.predict(test_input)
```

```python
from sklearn.metrics import accuracy_score, cohen_kappa_score, confusion_matrix

ConfusionMatrix = confusion_matrix(test_output, predicted_output)

Accuracy=accuracy_score(test_output, predicted_output)
```

```python
print("Confusion Matrix:\n", ConfusionMatrix)
print("\nAccuracy: ",Accuracy)
```
