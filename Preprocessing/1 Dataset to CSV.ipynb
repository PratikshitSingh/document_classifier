```python
import os
import pandas as pd
from sklearn.model_selection import train_test_split
data_folder=r"/media/pratikshit/6CBAB0B1BAB078E2/DIH/DIH/dataset/bbc"
folders=["business","politics","sports","entertainment","technology"]

os.chdir(data_folder)

x=[]
y=[]

for i in folders:
    files=os.listdir(i)
    for text_files in files:
        file_path=i+"/"+text_files
        with open(file_path, 'r', errors='replace') as f:
            data=f.readlines()
        data=''.join(data)
        x.append(data)
        y.append(i)
        
data={'news':x,'type':y}  
df=pd.DataFrame(data)

df.to_csv("newdatasets.csv",index=False)

df
```
